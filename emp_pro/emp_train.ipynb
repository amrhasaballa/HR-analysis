```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import xml.etree.ElementTree as ET
import sklearn.metrics
```

```python
train = pd.read_csv('hr_emp.csv')
train["Department"].unique()
```

```python
train1=train.drop(columns=['EmployeeCount','StandardHours','EmployeeNumber','Over18','Attrition'])
train1
```

```python
from sklearn.model_selection import train_test_split
train_df,test_df = train_test_split(train1, test_size=0.20, random_state=80)

```

```python


print('Total duplicate Values in the train dataset = ',train.duplicated())
print('Total missing values in train dataset = ',train.isnull().sum())
```

```python
qualitative =  [f for f  in train_df.columns if train.dtypes[f] == 'object']

qualitative
```

```python
for i in qualitative :
    
    train_df[i] = pd.Categorical(train_df[i])
    print(i,train_df[i].unique(),train_df[i].head(5))
    train_df[i] = pd.factorize(train_df[i])[0]
    print(i,train_df[i].unique(),train_df[i].head(5))
#train_df=pd.get_dummies(data=train_df,columns=qualitative)


```

```python
for i in qualitative :
    
    test_df[i] = pd.Categorical(test_df[i])
    test_df[i] = pd.factorize(test_df[i])[0]
```

```python
test_df.info()
```

```python
X_train = train_df.drop(columns=[ 'Department'])
y_train = train_df['Department']
```

```python
print (train['JobRole'].unique())
```

```python
X_test = train_df.drop(columns=[ 'Department'])
y_test = train_df['Department']
print(len(X_test))
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2 , f_regression

KBest = SelectKBest(score_func = f_regression, k = 6)
print(KBest)
KBest = KBest.fit(X_train,y_train)

df_scores = pd.DataFrame({'features': X_train.columns, 'f_regression': KBest.scores_ })
df_scores1=df_scores.sort_values(by='f_regression', ascending=False)
df_scores1
```

```python
df_scores2=df_scores1[:6]
selected_features=df_scores2['features']
df_scores2
```

```python
cols = KBest.get_support(indices=True)
features = X_train.columns[cols]
list(features)
```

```python
final_train = X_train[selected_features]

final_test=X_test[selected_features]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
#X, y = make_classification(n_samples=1000, n_features=4,n_informative=2, n_redundant=0,random_state=0, shuffle=False)
clf = RandomForestClassifier(max_depth=6)
#CV_rfc = GridSearchCV(estimator=rfc,cv= 10, scoring = 'roc_auc')
clf.fit(final_train ,y_train)
print(clf.predict_proba(final_test))
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
from sklearn import metrics
predictions = clf.predict(final_test)

#print(y_test.head(10))
#print(predictions[6])
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
#type(final_test)
print("accuracy: {:0.3f}".format(metrics.accuracy_score(y_test,predictions)))
```
