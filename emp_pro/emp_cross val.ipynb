```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import xml.etree.ElementTree as ET
import sklearn.metrics
```

```python
train = pd.read_csv('hr_emp.csv')
train["Department"].unique()
```

```python
train1=train.drop(columns=['EmployeeCount','StandardHours','EmployeeNumber','Over18'])
train1
```

```python
print('Total duplicate Values in the train dataset = ',train.duplicated())
print('Total missing values in train dataset = ',train.isnull().sum())
```

```python
qualitative =  [f for f  in train.columns if train.dtypes[f] == 'object']

qualitative
```

```python
for i in qualitative :
    
    train[i] = pd.Categorical(train[i])
    train[i] = pd.factorize(train[i])[0]
#train_df=pd.get_dummies(data=train_df,columns=qualitative)
train.head()
```

```python
X_train = train.drop(columns=[ 'Department'])
y_train = train['Department']
```

```python
 from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2 , f_regression

KBest = SelectKBest(score_func = f_regression, k = 6)
print(KBest)
KBest = KBest.fit(X_train,y_train)

df_scores = pd.DataFrame({'features': X_train.columns, 'f_regression': KBest.scores_ })
df_scores1=df_scores.sort_values(by='f_regression', ascending=False)
df_scores1
```

```python
df_scores2=df_scores1[:5]
selected_features=df_scores2['features']
df_scores2
```

```python
cols = KBest.get_support(indices=True)
features = X_train.columns[cols]
list(features)
```

```python
final_train = X_train[selected_features]
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=6)
```

```python
scoring = ['precision_macro', 'recall_macro','r2']
```

```python
scores = cross_validate(clf, final_train, y_train, scoring=scoring,cv=3)
```
